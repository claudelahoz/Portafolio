## Automóvil
Información del conjunto de datos
información adicional

Este conjunto de datos consta de tres tipos de entidades: (a) la especificación de un automóvil en términos de diversas características, (b) su clasificación de riesgo de seguro asignada, (c) sus pérdidas normalizadas en uso en comparación con otros automóviles. La segunda clasificación corresponde al grado en que el automóvil es más riesgoso de lo que indica su precio. A los automóviles se les asigna inicialmente un símbolo de factor de riesgo asociado con su precio. Luego, si es más riesgoso (o menos), este símbolo se ajusta moviéndolo hacia arriba (o hacia abajo) en la escala. Los actuarios llaman a este proceso "simbolización". Un valor de +3 indica que el automóvil es riesgoso, -3 que probablemente sea bastante seguro. El tercer factor es el pago de pérdidas promedio relativo por año de vehículo asegurado. Este valor está normalizado para todos los automóviles dentro de una clasificación de tamaño particular (pequeños de dos puertas, familiares, deportivos/especiales, etc.) y representa la pérdida promedio por automóvil por año. Nota: Varios de los atributos de la base de datos podrían usarse como un atributo de "clase".

```python
def cargar_imports_85(ruta_data, ruta_names):
    df = pd.read_csv(ruta_data, header=None)

    nombres_columnas = []
    with open(ruta_names, 'r') as file:
        capturar_nombres = False
        for line in file:
            if line.startswith('7. Attribute Information:'):
                capturar_nombres = True
                continue
            if capturar_nombres and line.strip().startswith(str(len(nombres_columnas) + 1) + '.'):
                nombre = line.split(':')[0].split('.')[1].strip()
                nombres_columnas.append(nombre)
            if len(nombres_columnas) == 26:  # Hay 26 atributos en total
                break
    
    df.columns = nombres_columnas
    df = df.replace('?', pd.NA)
    columnas_numericas = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 
                          'engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 
                          'city-mpg', 'highway-mpg', 'price']
    for col in columnas_numericas:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    
    return df

# Especifica la ruta de los archivos
ruta_base = "C:/Users/delahozce/OneDrive - GLOBAL HITSS/Documentos/Proyectos/Codes/git_portafolio/Portafolio/data/proyecto/httpsarchiveicsuciedu/Automovil/"  # Reemplaza esto con la ruta real donde están tus archivos
ruta_data = os.path.join(ruta_base, "imports-85.data")
ruta_names = os.path.join(ruta_base, "imports-85.names")

# Cargar los datos
car = cargar_imports_85(ruta_data, ruta_names)

print(car.shape)
print(car.info())
car.head(3)
```

La licencia de licor clase E, para tiendas de comestibles, licorerías, tiendas de conveniencia, etc., permite a los establecimientos comerciales vender licor para consumo fuera del local en envases originales sin abrir.

```python
import pandas as pd
import matplotlib.pyplot as plt

null_counts = Iowa_Liquor.isnull().sum()
null_percentages = 100 * Iowa_Liquor.isnull().sum() / len(Iowa_Liquor)

null_table = pd.concat([null_counts, null_percentages], axis=1, keys=['Número de nulos', 'Porcentaje de nulos'])
null_table = null_table.sort_values('Número de nulos', ascending=False)

print(null_table)

plt.figure(figsize=(12, 6))
null_table['Porcentaje de nulos'].plot(kind='bar')
plt.title('Porcentaje de valores nulos por columna')
plt.xlabel('Columnas')
plt.ylabel('Porcentaje de valores nulos')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
